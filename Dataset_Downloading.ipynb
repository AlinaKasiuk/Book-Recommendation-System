## Dataset Downloading

The goal of this Project is to create <u>an intelligent recommender of services or products</u>.

The dataset used in the Project is the Book-Crossing Dataset by the University of Freiburg that can be downloaded using the link: http://www.informatik.uni-freiburg.de/~cziegler/BX/

```python
!wget --no-check-certificate http://www2.informatik.uni-freiburg.de/~cziegler/BX/BX-CSV-Dump.zip
```

```python
import os
import zipfile

if not os.path.exists('./Datasets/'):
    os.makedirs('./Datasets/')

    # Unzip the archive
    local_zip = './BX-CSV-Dump.zip'
    zip_ref = zipfile.ZipFile(local_zip, 'r')
    zip_ref.extractall('./Datasets/')

    zip_ref.close()
```
